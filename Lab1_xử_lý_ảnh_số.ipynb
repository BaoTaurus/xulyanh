```python
# 1. Vit chng trnh np nh mu v hin th 3 mu khc nhau

import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Np nh mu
data = iio.imread('bird.png')

# Tch 3 knh mu
red = data[:, :, 0]    # Knh  (Red)
green = data[:, :, 1]  # Knh xanh l (Green)
blue = data[:, :, 2]   # Knh xanh dng (Blue)

# Hin th kt qu
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(red, cmap='gray')
plt.title('Knh  (Red)')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(green, cmap='gray')
plt.title('Knh Xanh L (Green)')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(blue, cmap='gray')
plt.title('Knh Xanh Dng (Blue)')
plt.axis('off')
plt.show()
```

```python
# 2. Vit chng trnh np nh v chuyn i gi tr knh Hue ca mu HSV, lu nh mi

import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
import colorsys

# Np nh mu
data = iio.imread('bird.png')

# Hm vector ha  p dng cho ton nh
rgb_to_hsv = np.vectorize(colorsys.rgb_to_hsv)
hsv_to_rgb = np.vectorize(colorsys.hsv_to_rgb)

# Chuyn i RGB sang HSV
h, s, v = rgb_to_hsv(data[:, :, 0] / 255.0, data[:, :, 1] / 255.0, data[:, :, 2] / 255.0)

# Thay i gi tr Hue mi = 1/3 Hue c
h_new = h / 3.0

# Chuyn ngc li RGB
r_new, g_new, b_new = hsv_to_rgb(h_new, s, v)
rgb_new = np.dstack((r_new, g_new, b_new)) * 255.0
rgb_new = rgb_new.astype(np.uint8)

# Lu v hin th
iio.imsave('bird_hue_new.png', rgb_new)
plt.imshow(rgb_new)
plt.title('nh vi Hue mi = 1/3 Hue c')
plt.axis('off')
plt.show()
```

```python
 # 3. Vit chng trnh np nh, chuyn i sang mu HSV, thay th gi tr Hue bng bnh phng gi tr Hue ban u, chuyn ngc li RGB v lu nh

 import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
import colorsys

# Np nh mu
data = iio.imread('bird.png')

# Hm vector ha  p dng cho ton nh
rgb_to_hsv = np.vectorize(colorsys.rgb_to_hsv)
hsv_to_rgb = np.vectorize(colorsys.hsv_to_rgb)

# Chuyn i RGB sang HSV
h, s, v = rgb_to_hsv(data[:, :, 0] / 255.0, data[:, :, 1] / 255.0, data[:, :, 2] / 255.0)

# Thay i gi tr Hue mi = bnh phng Hue c
h_new = h ** 2

# Chuyn ngc li RGB
r_new, g_new, b_new = hsv_to_rgb(h_new, s, v)
rgb_new = np.dstack((r_new, g_new, b_new)) * 255.0
rgb_new = rgb_new.astype(np.uint8)

# Lu v hin th
iio.imsave('bird_hue_squared.png', rgb_new)
plt.imshow(rgb_new)
plt.title('nh vi Hue mi = Hue c bnh phng')
plt.axis('off')
plt.show()
```
