```python
# 1. Viết chương trình nạp ảnh màu và hiển thị 3 màu khác nhau

import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Nạp ảnh màu
data = iio.imread('bird.png')

# Tách 3 kênh màu
red = data[:, :, 0]    # Kênh đỏ (Red)
green = data[:, :, 1]  # Kênh xanh lá (Green)
blue = data[:, :, 2]   # Kênh xanh dương (Blue)

# Hiển thị kết quả
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(red, cmap='gray')
plt.title('Kênh Đỏ (Red)')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(green, cmap='gray')
plt.title('Kênh Xanh Lá (Green)')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(blue, cmap='gray')
plt.title('Kênh Xanh Dương (Blue)')
plt.axis('off')
plt.show()
```

```python
# 2. Viết chương trình nạp ảnh và chuyển đổi giá trị kênh Hue của màu HSV, lưu ảnh mới

import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
import colorsys

# Nạp ảnh màu
data = iio.imread('bird.png')

# Hàm vector hóa để áp dụng cho toàn ảnh
rgb_to_hsv = np.vectorize(colorsys.rgb_to_hsv)
hsv_to_rgb = np.vectorize(colorsys.hsv_to_rgb)

# Chuyển đổi RGB sang HSV
h, s, v = rgb_to_hsv(data[:, :, 0] / 255.0, data[:, :, 1] / 255.0, data[:, :, 2] / 255.0)

# Thay đổi giá trị Hue mới = 1/3 Hue cũ
h_new = h / 3.0

# Chuyển ngược lại RGB
r_new, g_new, b_new = hsv_to_rgb(h_new, s, v)
rgb_new = np.dstack((r_new, g_new, b_new)) * 255.0
rgb_new = rgb_new.astype(np.uint8)

# Lưu và hiển thị
iio.imsave('bird_hue_new.png', rgb_new)
plt.imshow(rgb_new)
plt.title('Ảnh với Hue mới = 1/3 Hue cũ')
plt.axis('off')
plt.show()
```

```python
 # 3. Viết chương trình nạp ảnh, chuyển đổi sang màu HSV, thay thế giá trị Hue bằng bình phương giá trị Hue ban đầu, chuyển ngược lại RGB và lưu ảnh

 import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt
import colorsys

# Nạp ảnh màu
data = iio.imread('bird.png')

# Hàm vector hóa để áp dụng cho toàn ảnh
rgb_to_hsv = np.vectorize(colorsys.rgb_to_hsv)
hsv_to_rgb = np.vectorize(colorsys.hsv_to_rgb)

# Chuyển đổi RGB sang HSV
h, s, v = rgb_to_hsv(data[:, :, 0] / 255.0, data[:, :, 1] / 255.0, data[:, :, 2] / 255.0)

# Thay đổi giá trị Hue mới = bình phương Hue cũ
h_new = h ** 2

# Chuyển ngược lại RGB
r_new, g_new, b_new = hsv_to_rgb(h_new, s, v)
rgb_new = np.dstack((r_new, g_new, b_new)) * 255.0
rgb_new = rgb_new.astype(np.uint8)

# Lưu và hiển thị
iio.imsave('bird_hue_squared.png', rgb_new)
plt.imshow(rgb_new)
plt.title('Ảnh với Hue mới = Hue cũ bình phương')
plt.axis('off')
plt.show()
```
